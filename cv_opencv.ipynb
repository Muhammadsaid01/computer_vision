```python

import cv2
```

```python
from google.colab.patches import cv2_imshow
image = cv2.imread("/content/cat.jpg")
cv2_imshow(image)
```

```python
print("SHAPE:", image.shape)
# 455 => height
# 674 => width
# 3 => color channels (BRG) =>  Red, Green, Blue

print("HEIGHT:", image.shape[0])
print("WIDTH:", image.shape[1])
print("CHANNELS:", image.shape[2])
```

```python
(b ,g, r) = image[100, 50]
print("Blue:", b, "Green:", g, "Red:", r)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized = cv2.resize(gray, (400, 300))
cv2_imshow(resized)
```

```python
cropped = gray[100:320, 200:450]
cv2_imshow(cropped)
```

```python
rotated = cv2.rotate(gray, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated)
```

```python
flipped = cv2.flip(gray, 1)
cv2_imshow(flipped)
```

```python
cv2.imwrite("cropped_cat.jpeg", cropped)
```

# OpenCV Video Codes


```python
# import cv2
import cv2
cap = cv2.VideoCapture(0)
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

while True:
    success, frame = cap.read()  # True, False | Boolean
    if not success:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    face = face_cascade.detectMultiScale(gray, 1.5, 4)
    for (x, y, w, h) in face:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 4)

    cv2.imshow("Face Detection", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):  # 'q' tugmasi bosilganda ochishi
        break

cap.release()  # Ochiq qolgan oynalrni yopish uchun
cv2.destroyAllWindows()

print("Hello World!")


url = "rtsp://192.168.1.164:8001/videp"
 cap = cv2.VideoCapture(url)
while True:
      success, frame = cap.read()  # True, False | Boolean
    if not success:
        break

#
#     gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
#     blurred = cv2.GaussianBlur(frame, (15, 15), 0)  # Rasmni blurlash
#     edgs = cv2.Canny(frame, 100, 200)  # Rasmning chekkalarini aniqlash
#     flip = cv2.flip(frame, 1)  # Rasmni gorizontal ravishda teskari qilish

#     # Oynani ochish va video oqimini ko'rsatish uchun
#     # cv2.imshow("Origina:", frame)
#     # cv2.imshow("Gray:", gray)
#     # cv2.imshow("Blurred:", blurred)
#     # cv2.imshow("Edges:", edgs)
#     cv2.imshow("Flipped:", flip)

#     if cv2.waitKey(1) & 0xFF == ord('q'):  # 'q' tugmasi bosilganda
#         break


# cap.release()
# cv2.destroyAllWindows()


# ==========================================#
```
